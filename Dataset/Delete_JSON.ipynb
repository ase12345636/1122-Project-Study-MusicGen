```python


def delete_json(music_path: str,json_path: str):
    music_items = os.listdir(music_path)
    json_items = os.listdir(json_path)

    det_list = []
    for i in json_items :
        file=os.path.splitext(i)[0]
        if file not in music_items:
            det_list.append(i)

    print(f"Total Number of Delete Files: {len(det_list)}")
    print("---------- Start Delete ----------")

    for i in det_list:
        print(f"Delete {i} file")
        os.remove(json_path+i)
    print("---------- Delete Done ----------")

music_path = 'OriginalData/music/validation/'
json_path = 'OriginalData/json/validation/'
delete_json(music_path,json_path)
```
